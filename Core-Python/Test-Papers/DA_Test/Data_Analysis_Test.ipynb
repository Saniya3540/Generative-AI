## Q.1. Numerical Operations

```python
import numpy as np
```

```python
units_sold=np.array([10,25,40,35,50])
units_sold
```

```python
unit_price=np.array([100,150,200,120,180])
unit_price
```

```python
total_sales=units_sold*unit_price
total_sales
```

```python
overall_sales=np.sum(total_sales)
overall_sales
```

## Q.2. Indexing and Filtering

```python
#Using NumPy arrays
units_sold=np.array([10,25,40,35,50])
profit = np.array([200, 300, 500, 400, 700])
```

```python
result = units_sold[units_sold > 30]
result
```

```python
profit = profit[units_sold > 30]
profit
```

```python
avg_profit=np.mean(profit)
avg_profit
```

```python
import pandas as pd
```

```python
#Using Pandas

data={
    "Units_Sold": [10, 25, 40, 35, 50],
    "Unit_Price": [100, 150, 200, 120, 180],
    "Profit": [200, 300, 500, 400, 700]
}
data
```

```python
df=pd.DataFrame(data)
```

```python
df.head()
```

```python
df.dtypes
```

```python
result=df[df["Units_Sold"]>30]
result
```

```python
avg_profit=result["Profit"].mean()
avg_profit
```
